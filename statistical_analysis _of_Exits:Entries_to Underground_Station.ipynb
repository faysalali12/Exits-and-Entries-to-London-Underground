 # Part 1.1: Calculating Morning and Evening Peak Proportions
 
    

The overall aims of this coursework was to:

```python
import pandas as pd
import matplotlib as plt
import seaborn as sn
%matplotlib inline
```

```python
exits = pd.read_csv('hourlyExits.csv',index_col='Station')
exits
```

The table above shows the overall results for the hourly exits at all of the London Underground stations

```python
Peak_Counts= exits.copy()
Peak_Counts
Peak_Counts= Peak_Counts.assign( AM_Peak = Peak_Counts.H07 + Peak_Counts.H08 + Peak_Counts.H09, PM_Peak =Peak_Counts.H16 + Peak_Counts.H17 + Peak_Counts.H18)
Peak_Counts = Peak_Counts.assign(Total_Exits = Peak_Counts.apply(sum,axis=1))
Peak_Counts=Peak_Counts.assign(AM_Proportion_Of_Daily= Peak_Counts.AM_Peak/ Peak_Counts.Total_Exits)
Peak_Counts=Peak_Counts.assign(PM_Proportion_Of_Daily= Peak_Counts.PM_Peak / Peak_Counts.Total_Exits)
Peak_Counts
```

```python
Peak = pd.DataFrame()
Peak= Peak.assign(AM_Peak=Peak_Counts.AM_Peak, PM_Peak =Peak_Counts.PM_Peak , Total_Exits = Peak_Counts.Total_Exits ,AM_Proportion_Of_Daily = Peak_Counts.AM_Proportion_Of_Daily, PM_Proportion_Of_Daily = Peak_Counts.PM_Proportion_Of_Daily )
Peak
```

The AM_Peak in the table above represents the total number of exits during the AM peak time. The AM_Peak was calculated by finding the sum of the number of exits during the hours 07:00 (H07),08:00 (H08) and 09:00 (H09). The calculation for the results in the PM_Peak column were calculated using the number of exits during the hours 16:00 (H16),17:00 (H17) and 18:00 (H18). The total exits represents the overall exits occuring in the 24 hour period across all station in the London Underground. The AM_Proportion_of_Daily represents a ratio of the proportion of AM_peak exits and the total number of exits and the PM_proportion_Of_daily represents a ratio of the proportion of PM_peak exits and the total number of exits.

# Part 1.3: A Simple Classification of Stations

```python
def ClassifactionCategory(column):
    if (column.AM_Proportion_Of_Daily - column.PM_Proportion_Of_Daily >0.01  ) : return 'Work'
    if (column.AM_Proportion_Of_Daily - column.PM_Proportion_Of_Daily < 0.01) : return 'Residential'
    if (column.AM_Proportion_Of_Daily == column.PM_Proportion_Of_Daily) : return 'Other'

Peak.apply(ClassifactionCategory, axis=1)

Peak = Peak.assign(Classification = Peak.apply(ClassifactionCategory, axis=1))
Peak

```

The above table shows data for all station exits and includes: 
- The number of exits during the AM peak times
- PM peak times, the total number of exits 
- The proportion of AM exits with respect to the total exits
- The proportion of PM exits with respect to the total exits
- the classification of the stations based on the number of AM and PM peak exits

```python
Northern_line_class = Peak[["Classification","AM_Peak","PM_Peak","Total_Exits","AM_Proportion_Of_Daily","PM_Proportion_Of_Daily"]].loc[['Archway', 'Balham', 'Embankment', 'Goodge Street', 'Highgate', 'High Barnet', 
                                                  'Leicester Square', 'Morden', 'South Wimbledon', 'Tottenham Court Road', 'Warren Street']]
Northern_line_class
```

The table above shows the same data as the previous table but only for the stations along the Northern line.

The results produced in the Northern line table did produce an expected result. The stations that were classed as 'Work' are mainly stations that are located towards the centre of London. People who tend to use these stations during the AM and PM peak times do tend to use them to commute to work. Some of the other stations were mainly classed as residential areas and some are geographically located in residential areas of London. However, the station 'Leicester Square' was classed as a residential area even though it is in the centre of London. This error could be due to the threshold used when classing the different stations.
